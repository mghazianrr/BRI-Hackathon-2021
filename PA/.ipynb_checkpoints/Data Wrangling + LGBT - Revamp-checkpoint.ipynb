**Imma try to encode objects and fills null values**

```python
import pandas as pd
import numpy as np
import statistics
from sklearn.preprocessing import StandardScaler, LabelEncoder
import warnings
from bayes_opt import BayesianOptimization
from lightgbm import LGBMClassifier
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split as tts
from sklearn.model_selection import cross_val_score as cvs
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train.info()
train.head(10)
```

```python
test.info()
test.head(10)
```

```python
#set age
train['age'] = 2021 - train['age']
test['age'] = 2021 - test['age']
train.loc[0:10, "age"]
```

```python
train["year_graduated"] = 2021 - train["year_graduated"]
test["year_graduated"] = 2021 - test["year_graduated"]
train["year_graduated"].head(25)
```

```python
#Fill Last_achievement_%
la_mean = train["Last_achievement_%"].mean()
la_median = train["Last_achievement_%"].median()
la_mode = statistics.mode(train["Last_achievement_%"].dropna())
print(la_mean, la_median, la_mode)
train["Last_achievement_%"] = train["Last_achievement_%"].replace(np.NaN, la_mean)
```

```python
train["Last_achievement_%"].isnull().sum()
```

```python
#Fill Achievement_above_100%_during3quartal
aa_mean = train["Achievement_above_100%_during3quartal"].mean()
aa_median = train["Achievement_above_100%_during3quartal"].median()
aa_mode = statistics.mode(train["Achievement_above_100%_during3quartal"].dropna())
print(aa_mean, aa_median, aa_mode)
train["Achievement_above_100%_during3quartal"] = train["Achievement_above_100%_during3quartal"].replace(np.NaN, aa_median)
```

```python
train["Achievement_above_100%_during3quartal"].isnull().sum()
```

```python
#Object data type treatment
le = LabelEncoder()
#every cols
obj_cols = ['job_level', 'person_level', 'Employee_type', 'marital_status_maried(Y/N)', 'Education_level']
for col in obj_cols:
    labels = train[col].unique().tolist() + test[col].unique().tolist()
    le.fit(labels)
    test[col] = le.transform(test[col].astype(str))
    train[col] = le.transform(train[col].astype(str))
    print(list(le.classes_))
```

```python
X = train.drop('Best Performance', axis = 1)
y = train['Best Performance']
model = xgb.XGBRFClassifier(n_estimators=1000, random_state=1245)
score = cvs(model, X, y, cv=5, scoring="roc_auc", n_jobs=-1)
print("XGB ROC-AUC Mean Score: ", np.mean(score))
```

```python
warnings.filterwarnings('ignore')

dtrain = xgb.DMatrix(X, y, feature_names=X.columns.values)
def hyp_xgb(max_depth, subsample, colsample_bytree,min_child_weight, gamma, learning_rate):
    params = {
    'objective': 'binary:logistic',
    'eval_metric':'auc',
    'nthread':-1
     }
    
    params['max_depth'] = int(round(max_depth))
    params['subsample'] = max(min(subsample, 1), 0)
    params['colsample_bytree'] = max(min(colsample_bytree, 1), 0)
    params['min_child_weight'] = int(min_child_weight)
    params['gamma'] = max(gamma, 0)
    params['learning_rate'] = learning_rate
    
    scores = xgb.cv(params, dtrain, num_boost_round=500,verbose_eval=False, 
                    early_stopping_rounds=10, nfold=5)
    return scores['test-auc-mean'].iloc[-1]

pds ={
  'min_child_weight':(3, 20),
  'gamma':(0, 10),
  'subsample':(0.5, 1),
  'colsample_bytree':(0.1, 1),
  'max_depth': (2, 15),
  'learning_rate': (0.01, 0.5)
}
```

```python
optimizer = BayesianOptimization(hyp_xgb, pds, random_state=1)
optimizer.maximize(init_points=4, n_iter=25)
```

```python
optimizer.max['params']
```

```python
params = {
    'colsample_bytree': 0.9548842927260913,
    'gamma': 0.2027156134070207,
    'learning_rate': 0.03691008545891213,
    'max_depth': 13,
    'min_child_weight': 19.575340276597952,
    'subsample': 0.7351913053596237,
    'objective': 'binary:logistic',
    'eval_metric':'auc',
    'n_jobs':-1
}

xgbr =  xgb.XGBClassifier(**params, random_state=12345, nthread=-1)
xgbr.fit(X, y)
```

```python
pred = xgbr.predict_proba(test)[:,1]
```

```python
print(pred)
```

```python
submission = pd.read_csv("sample_submission.csv")
submission['Best Performance'] = pred
```
